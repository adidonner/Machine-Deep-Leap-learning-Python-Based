USE KERS a statistical analysis for the Fashion MNIST data set. How many items are there from each group?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

# Load the Fashion MNIST dataset
fashion_mnist = fetch_openml('Fashion-MNIST', version=1)
data = fashion_mnist.data
labels = fashion_mnist.target.astype(int)

# Combine data and labels into a DataFrame
df = pd.DataFrame(data, columns=[f'pixel_{i}' for i in range(data.shape[1])])
df['label'] = labels

# Split the data into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Count the number of items in each group
train_counts = train_df['label'].value_counts()
test_counts = test_df['label'].value_counts()

# Display the counts
print("Training Set:")
print(train_counts)
print("\nTesting Set:")
print(test_counts)

# Visualize the distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

train_counts.plot(kind='bar', ax=axes[0], color='blue')
axes[0].set_title('Training Set Distribution')

test_counts.plot(kind='bar', ax=axes[1], color='green')
axes[1].set_title('Testing Set Distribution')

plt.show()
```

Print the probability of getting an object from each category and find what is the maximum

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

# Load the Fashion MNIST dataset
fashion_mnist = fetch_openml('Fashion-MNIST', version=1)
data = fashion_mnist.data
labels = fashion_mnist.target.astype(int)

# Map category names to corresponding indices
category_names = {
    0: 'T-shirt/top',
    1: 'Trouser',
    2: 'Pullover',
    3: 'Dress',
    4: 'Coat',
    5: 'Sandal',
    6: 'Shirt',
    7: 'Sneaker',
    8: 'Bag',
    9: 'Ankle boot'
}

# Combine data and labels into a DataFrame
df = pd.DataFrame(data, columns=[f'pixel_{i}' for i in range(data.shape[1])])
df['label'] = labels

# Map category names to the DataFrame
df['category'] = df['label'].map(category_names)

# Split the data into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Calculate the probability of each category
total_train_items = len(train_df)
train_probabilities = train_df['category'].value_counts() / total_train_items

total_test_items = len(test_df)
test_probabilities = test_df['category'].value_counts() / total_test_items

# Display the probabilities
print("Training Set Probabilities:")
print(train_probabilities)
print("\nTesting Set Probabilities:")
print(test_probabilities)

# Find the category with the maximum probability
max_train_category = train_probabilities.idxmax()
max_test_category = test_probabilities.idxmax()

print("\nCategory with Maximum Probability in Training Set:", max_train_category)
print("Category with Maximum Probability in Testing Set:", max_test_category)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from scipy.special import softmax

# Load the Fashion MNIST dataset
fashion_mnist = fetch_openml('Fashion-MNIST', version=1)
data = fashion_mnist.data
labels = fashion_mnist.target.astype(int)

# Map category names to corresponding indices
category_names = {
    0: 'T-shirt/top',
    1: 'Trouser',
    2: 'Pullover',
    3: 'Dress',
    4: 'Coat',
    5: 'Sandal',
    6: 'Shirt',
    7: 'Sneaker',
    8: 'Bag',
    9: 'Ankle boot'
}

# Combine data and labels into a DataFrame
df = pd.DataFrame(data, columns=[f'pixel_{i}' for i in range(data.shape[1])])
df['label'] = labels

# Map category names to the DataFrame
df['category'] = df['label'].map(category_names)

# Split the data into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Calculate the softmax probabilities for each category
train_probabilities = softmax(train_df['label'].value_counts().sort_index(), axis=0)
test_probabilities = softmax(test_df['label'].value_counts().sort_index(), axis=0)

# Display the softmax probabilities
print("Training Set Softmax Probabilities:")
print(train_probabilities)
print("\nTesting Set Softmax Probabilities:")
print(test_probabilities)

# Find the category with the maximum softmax probability
max_train_category = category_names[np.argmax(train_probabilities)]
max_test_category = category_names[np.argmax(test_probabilities)]

print("\nCategory with Maximum Softmax Probability in Training Set:", max_train_category)
print("Category with Maximum Softmax Probability in Testing Set:", max_test_category)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from scipy.special import softmax

# Load the Fashion MNIST dataset
fashion_mnist = fetch_openml('Fashion-MNIST', version=1)
data = fashion_mnist.data
labels = fashion_mnist.target.astype(int)

# Map category names to corresponding indices
category_names = {
    0: 'T-shirt/top',
    1: 'Trouser',
    2: 'Pullover',
    3: 'Dress',
    4: 'Coat',
    5: 'Sandal',
    6: 'Shirt',
    7: 'Sneaker',
    8: 'Bag',
    9: 'Ankle boot'
}

# Combine data and labels into a DataFrame
df = pd.DataFrame(data, columns=[f'pixel_{i}' for i in range(data.shape[1])])
df['label'] = labels

# Map category names to the DataFrame
df['category'] = df['label'].map(category_names)

# Split the data into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Calculate the softmax probabilities for each category
train_probabilities = softmax(train_df['label'].value_counts().sort_index(), axis=0)
test_probabilities = softmax(test_df['label'].value_counts().sort_index(), axis=0)

# Display the softmax probabilities with category names
print("Training Set Softmax Probabilities:")
for category, probability in zip(category_names.values(), train_probabilities):
    print(f"{category}: {probability:.4f}")

print("\nTesting Set Softmax Probabilities:")
for category, probability in zip(category_names.values(), test_probabilities):
    print(f"{category}: {probability:.4f}")

# Find the category with the maximum softmax probability
max_train_category = category_names[np.argmax(train_probabilities)]
max_test_category = category_names[np.argmax(test_probabilities)]

print("\nCategory with Maximum Softmax Probability in Training Set:", max_train_category)
print("Category with Maximum Softmax Probability in Testing Set:", max_test_category)
```
